```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
regData = np.loadtxt("./resources/RegData.csv", delimiter = ",")
```

```python
X = regData[:,0]
y = regData[:,1]
```

```python
plt.scatter(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
plt.scatter(X_train, y_train, color='red', label = 'Training Data')
plt.scatter(X_test, y_test, color='g', label = 'Testing Data')
plt.legend()
```

```python
LR = LinearRegression()
LR.fit(X_train.reshape(-1, 1), y_train)
```

```python
predection = LR.predict(X_test.reshape(-1, 1))
```

```python
plt.plot(X_test, predection)
plt.scatter(X_test, y_test, color = 'y')
```

```python
LR.score(X_test.reshape(-1, 1), y_test)
```
