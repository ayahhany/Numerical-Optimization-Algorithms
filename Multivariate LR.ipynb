```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
multiple = np.loadtxt("./resources/MultipleLR.csv", delimiter = ",")
```

```python
X0 = np.ones(len(multiple))
X1 = multiple[:, 0]
X2 = multiple[:,1]
X3 = multiple[:,2]
y = multiple[:,3]
y = y.reshape(25,1)
```

```python
def multivariate_LR(x0,x1,x2,x3,y,alpha):
    #1. assume thetas = 0
    thetas = np.zeros(4).reshape(4,1)
    thetas_lst=np.zeros((100,4))
    js =np.zeros(100)


    #2. x matrix
    X = np.matrix([x0,x1,x2,x3]).T
    for i in range(100):
        #if i> 0 : print(gradients.shape)
        #3. calculate the hypothesis
        hx = (np.dot(X,thetas)).reshape(25,1)
        
        #4. calculate loss/cost function
        e = (hx - y).reshape(25,1)
        #print(e)
        j =(1/(len(y)*2))* np.sum(np.square(e))
        
        thetas_lst[i,:]=thetas.reshape(1,4)
        js[i] = j
      
        #gradient=((np.dot(x.T,e))/m).reshape(n,1)

        #5. calculate the gradient
        gradients = ((np.dot(X.T, e))/len(y)).reshape(4,1)
        
        #6. Update the parameters (simultaneously)
        thetas = thetas - alpha * gradients
    js = js.reshape(25,4)
    return thetas_lst, js
```

```python
thetas, js = multivariate_LR(X0,X1,X2,X3,y,0.0001)
```

```python
print(f'''the min cost is {js[-1]}''')
```

```python
plt.plot(np.arange(len(js)), js,'-o')
plt.xlabel("Epochs")
plt.ylabel("J")
plt.title("Epochs vs Cost function")
```
